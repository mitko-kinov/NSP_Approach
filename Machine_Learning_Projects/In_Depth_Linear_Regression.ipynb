## Polynomial Basis Functions

```python
# Importing libraries:

from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
# Packed example part1:

x = np.array([2, 3, 4])
poly = PolynomialFeatures(3, include_bias = False)
poly.fit_transform(x[:, None])
```

```python
# Packed example part2:

poly_model = make_pipeline(PolynomialFeatures(7),
                           LinearRegression())
```

```python
# Packed example part3:

rng = np.random.RandomState(1)

x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)

xfit = np.linspace(0, 10, 1000)
poly_model.fit(x[:, np.newaxis], y)
yfit = poly_model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
```

## Gaussian Basis Functions:

```python
# Importing libraries:

from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class GaussianFeatures(BaseEstimator, TransformerMixin):
    '''Uniformly spread Gaussian features for one-dimensional input'''
    
    def __init__(self, N, width_factor = 2.0):
        self.N = N
        self.width_factor = width_factor
        
    @staticmethod
    def _gauss_basis(x, y, width, axis = None):
        arg = (x - y) / width
        return np.exp(-0.5 * np.sum(arg ** 2, axis))
    
    def fit(self, X, y = None):
        # create N centes spread along the data range
        self.centers_ = np.linspace(X.min(), X.max(), self.N)
        self.width_ = self.width_factor * (self.centers_[1] - self.centers_[0])
        return self
    
    def transform(self, X):
        return self._gauss_basis(X[:, :, np.newaxis], self.centers_, self.width_, axis = 1)
```

```python
# Building GaussianFeatures and LinearRegression pipeline:

gauss_model = make_pipeline(GaussianFeatures(20),
                            LinearRegression())
```

```python
# Fitting the model:

gauss_model.fit(x[:, np.newaxis], y)
```

```python
# Predicting with the model:

yfit = gauss_model.predict(xfit[:, np.newaxis])
```

```python
# Plotting the results:

plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.xlim(0, 10)
```

## Regularization:

```python
# Overfitting with Gaussian basis functions:

model = make_pipeline(GaussianFeatures(30),
                      LinearRegression())

yfit = model.fit(x[:, np.newaxis], y).predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.xlim(0, 10)
plt.ylim(-4.5, 1.5)
```

```python
# Building a function for plotting the coef. of the Gaussina basis with restpect to their locations:

def basis_plot(model, title = None):
    fig, ax = plt.subplots(2, sharex = True)
    model.fit(x[:, np.newaxis], y)
    yfit = model.predict(xfit[:, np.newaxis])
    ax[0].scatter(x, y)
    ax[0].plot(xfit, yfit)
    ax[0].set(xlabel = 'x', ylabel = 'y', ylim = (-1.5, 1.5))
    
    if title:
        ax[0].set_title(title)
        
    ax[1].plot(model.steps[0][1].centers_,
               model.steps[1][1].coef_)
    ax[1].set(xlabel = 'Basis Location',
              ylabel = 'Coefficient',
              xlim = (0, 10))
```

```python
# Activating the function to see results:

model = make_pipeline(GaussianFeatures(30), LinearRegression())
basis_plot(model)
```

## Rigid Regression (L2 Regularization):

```python
# Importing libraries:

from sklearn.linear_model import Ridge
```

```python
# Building GaussianFeatures and Ridge pipeline:

model = make_pipeline(GaussianFeatures(30), Ridge(alpha = 0.1))
```

```python
# Activating the function to see results with Ridge:

basis_plot(model, title = 'Ridge Regression')
```
